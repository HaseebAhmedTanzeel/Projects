# DATA MINING

## Importing Libraries

```python
import pandas as pd
import numpy as np
from requests import get

#library that makes it easy to scrape information from web pages. It sits atop an HTML or XML parser, providing Pythonic idioms for iterating, searching, and modifying the parse tree
from bs4 import BeautifulSoup as bs
```

## Initiating variables for data mining

```python
urls = []
typeList = []
priceList = []
locationList = []
bathsList = []
bedsList = []
areaList = []
```

```python
def my_func(soup):
    main_list = soup.find('ul', {"class":"_357a9937"})
    for i in main_list.find_all('li', {"class":"ef447dde"}):
        div = i.find('div',{"class":"f74e80f3"})
        link = div.find('a', href = True)
        link = 'https://www.zameen.com/'+link['href']
        urls.append(link)
```

```python
for i in range(1,850):
    if i%50==0:
        print(i)
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'}
        url = 'https://www.zameen.com/Rentals/Lahore-1-'+str(i)+'.html?sort=date_desc'
        response = get(url, headers=headers)
        soup = bs(response.text, 'html.parser')
        my_func(soup)
    except:
        print(i,'Connection forcibly closed by remote host')
```

```python
urls
```

```python
n = 0
for i in urls:
    print(n)
    try:
        #time.sleep(30)
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'}
        response = get(i, headers=headers)
        #print("response: ",response)
        soup = bs(response.text, 'html.parser')
        overview = soup.find('div', {"class":"_066bb126"}).find('ul', {"class":"_033281ab"}).find_all('li')
        #print("overview: ",overview)
        for i in overview:
            if i.find("span", class_="_812aa185")["aria-label"] == 'Type':
                typeList.append(i.find("span", class_="_812aa185").text)
            if i.find("span", class_="_812aa185")["aria-label"] == 'Price':
                priceList.append(i.find("span", class_="_812aa185").text)
            if i.find("span", class_="_812aa185")["aria-label"] == 'Location':
                locationList.append(i.find("span", class_="_812aa185").text)
            if i.find("span", class_="_812aa185")["aria-label"] == 'Area':
                areaList.append(i.find("span", class_="_812aa185").text)
            if i.find("span", class_="_812aa185")["aria-label"] == 'Baths':
                bathsList.append(i.find("span", class_="_812aa185").text)
            if i.find("span", class_="_812aa185")["aria-label"] == 'Beds':
                bedsList.append(i.find("span", class_="_812aa185").text)
    except:
        print("\n",response)
        print("\nException for : "+i)
    n+=1
```

```python
df = pd.DataFrame(list(zip(typeList,priceList, locationList, bathsList, areaList, bedsList)),
                 columns = ['Type', 'Price', 'Location', 'Baths', 'Area', 'Beds'])
df
```

## Formatting data into the needed data types and standardizing it

```python
# Removing PKR from price since the website is solely of Pakistan we do not need to make changes to this part other than removing currency name
df['Price'] = df['Price'].str.split('PKR').str.get(1)
df
```

```python
# We need to convert the price into numeric/integer value hence we move the string thousand and store its numeric equivalent 1000 in Price Unit and similarly do it with lakh
df['Price Unit'] = df['Price'].apply(lambda x: '1000' if 'Thousand' in x else '100000')

# Keeping only numbers and removing the words
df['Price'] = df['Price'].str.split(' ').str.get(0)
df
```

```python
# Converting price to float as we had values such as 1.1 lac. If we had used integer we would have lost them.
df['Price'] = df['Price'].astype(float)

# Converting Price Unit to integer
df['Price Unit'] = df['Price Unit'].astype(int)

# Multiplying them to get final rental value as integer since we know we can't have rents as fractional value i.e paisas (which are discontinued)
df['Price'] = df['Price'] * df['Price Unit']
df['Price'] = df['Price'].astype(int)

#Dropping price unit since we no longer need it
df.drop(['Price Unit'], axis=1, inplace=True)
df
```

```python
# Similar to price we have to remove Marla/Kanal and convert to lower unit of Area i.e Marla

# Store the unit of Area in Unit Area
df['Area Unit'] = df['Area'].str.split(' ').str.get(1)

# Keeping the numeric value
df['Area'] = df['Area'].str.split(' ').str.get(0)
df
```

```python
# Converting Area values to numeric and any errors are handled by coercing them to NaN
df['Area'] = pd.to_numeric(df['Area'], errors='coerce')

# Replace NaN values with zero
df['Area'] = df['Area'].fillna(0)

# Replace non-'Marla' or non-'Kanal' values in 'Area Unit' with 'Marla'
df['Area Unit'] = df['Area Unit'].replace({'Marla': 'Marla', 'Kanal': 'Kanal'}, regex=True).fillna('Marla')

# We multiply 
df['Area'] = df.apply(lambda x: x.Area*20 if 'Kanal' in x['Area Unit'] else x['Area'], axis =1)
df.drop(['Area Unit'], axis=1, inplace=True)
df
```

```python
# Convert 'Beds', and 'Baths' columns to numeric and set non-numeric values to NaN
df['Beds'] = pd.to_numeric(df['Beds'], errors='coerce')
df['Baths'] = pd.to_numeric(df['Baths'], errors='coerce')

# Replacing NaN values with zero
df['Beds'] = df['Beds'].fillna(0)
df['Baths'] = df['Baths'].fillna(0)

# 'Baths' and 'Beds' column set to type integer
df[['Baths', 'Beds']] = df[['Baths', 'Beds']].astype(int)
df
```

```python
# Removing city, Pakistan because we are currently dealing with only one city
df['Location'] = df['Location'].replace(', Lahore, Punjab', '', regex=True)
df
```

```python
# We will use this later on to check how much data we lost due to cleaning
original_length_of_dataframe = len(df)
original_length_of_dataframe
```

## Data cleaning and finding Outliers

```python
# Create a box plot for 'Area'
plt.figure(figsize=(18, 5))
sns.boxplot(x=df['Area'])
plt.title('Box Plot of Area')
plt.xlabel('Area')
#plt.xticks(range(0, len(df['Area']), 10))  # Adjust the step value (10 in this example) to control tick frequency
plt.show()

# Create a box plot for 'Price'
plt.figure(figsize=(18, 5))
sns.boxplot(x=df['Price'])
plt.title('Box Plot of Price')
plt.xlabel('Price')
#plt.xticks(range(0, len(df['Price']), 10))  # Adjust the step value (10 in this example) to control tick frequency
plt.show()
```

```python
# Calculate the first quartile (Q1) and third quartile (Q3) for 'Area' and 'Price'
Q1_area = df['Area'].quantile(0.25)
Q3_area = df['Area'].quantile(0.75)
Q1_price = df['Price'].quantile(0.25)
Q3_price = df['Price'].quantile(0.75)

# Calculate the interquartile range (IQR) for 'Area' and 'Price'
IQR_area = Q3_area - Q1_area
IQR_price = Q3_price - Q1_price

# Calculate the lower and upper bounds for outliers for 'Area' and 'Price'
lower_bound_area = Q1_area - 1.5 * IQR_area
upper_bound_area = Q3_area + 1.5 * IQR_area
lower_bound_price = Q1_price - 1.5 * IQR_price
upper_bound_price = Q3_price + 1.5 * IQR_price

print("Area Outlier Thresholds:")
print("Lower Bound:", lower_bound_area)
print("Upper Bound:", upper_bound_area)

print("\nPrice Outlier Thresholds:")
print("Lower Bound:", lower_bound_price)
print("Upper Bound:", upper_bound_price)
```

###### While the outliers are statistically right any subject matter expert or any person with general knowledge about Pakistan's real estate market knows that rentals and Area can not be negative. 
###### Usually the smallest available size in residential areas is 2.5 to 3 Marla. While there are some flats such as studio flats that can be smaller  so, a safe lower bound would be 1 marla.
###### Furthermore Residential Areas usually do not cross above 4 kanal. If it is a corner plot or category plot then it may be a bit more extra in size but that is rare.  Hence, a safe amount of upper bound can be 6 kanal which is 120 marla.
###### We do know rentals can not go negative but they can also not go below 1000 PKR per month also
###### The rental values can even go 5 lakhs. Hence a safe upper bound price would be 10 lakh or 1 million

```python
lower_bound_area = 1
upper_bound_area = 120

lower_bound_price = 1000
upper_bound_price = 1000000
```

```python
# Remove outliers from the 'Area' and 'Price' columns and saving them in a different dataframe
outliers_area = df[(df['Area'] < lower_bound_area) | (df['Area'] > upper_bound_area)]
outliers_price = df[(df['Price'] < lower_bound_price) | (df['Price'] > upper_bound_price)]
df = df[~df.index.isin(outliers_area.index) & ~df.index.isin(outliers_price.index)]
df
```

```python
# Checking removed data
print(outliers_area)
print(outliers_price)
```

```python
#Filter and remove records where 'Beds' and 'Baths' are zero
df = df[~((df['Beds'] == 0) | (df['Baths'] == 0))]

# Reset the DataFrame
df.reset_index(drop=True, inplace=True)
df
```

```python
data_loss = (1 - (len(df)/original_length_of_dataframe))*100
data_loss
```

```python
df['Price_to_Area_ratio'] = df['Price']/df['Area']
df
```

# Data Visulization

## By Location

### Only Location

```python
# Group by 'Location' and perform aggregations
df_location = df.groupby('Location').agg({
    'Price': ['min', 'mean', 'max'],
    'Beds': ['min', 'max'],
    'Area': ['min','mean','max'],
    'Price_to_Area_ratio' : ['min', 'mean', 'max'],
    'Location': 'count'
}).reset_index()

# Flatten the multi-level column index
df_location.columns = [
    'Location', 'Min_Price', 'Mean_Price', 'Max_Price', 
    'Min_Beds', 'Max_Beds', 'Min_Area', 'Mean_Area', 'Max_Area', 'Min_Price_to_Area_ratio','Mean_Price_to_Area_ratio',
    'Max_Price_to_Area_ratio', 'Num_of_Records'
]

df_location
```

```python
top_locations = df_location.sort_values(by='Num_of_Records', ascending=False).head(10).reset_index(drop=True)
top_locations
```

### Location and then Type

```python
# Group by 'Location' and perform aggregations
df_location_type = df.groupby(['Location', 'Type']).agg({
    'Price': ['min', 'mean', 'max'],
    'Beds': ['min', 'max'],
    'Area': ['min','mean','max'],
    'Price_to_Area_ratio' : ['min', 'mean', 'max'],
    'Location': 'count'
}).reset_index()

# Flatten the multi-level column index
df_location_type.columns = [
    'Location','Type' ,'Min_Price', 'Mean_Price', 'Max_Price', 
    'Min_Beds', 'Max_Beds', 'Min_Area', 'Mean_Area', 'Max_Area', 'Min_Price_to_Area_ratio','Mean_Price_to_Area_ratio',
    'Max_Price_to_Area_ratio', 'Num_of_Records'
]

df_location_type
```
